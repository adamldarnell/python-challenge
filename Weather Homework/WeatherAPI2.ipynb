```python
import requests as req
import matplotlib
from citipy import citipy
import pandas as pd
import numpy as np
```

```python
weatherkey = "9789f91d6357230d3b0d947d90b119cc"
```

```python
rlat = np.random.uniform(-90,90)
rlong = np.random.uniform(-180, 180)
```

```python
url = "http://api.openweathermap.org/data/2.5/"
query_url = url + "weather?lat=" + str(rlat) + "&lon=" + str(rlong) + "&appid=" + weatherkey
```

```python
location_data = pd.DataFrame()

location_data['random_lat'] = [np.random.uniform(-90,90) for x in range(500)]
location_data['random_lng'] = [np.random.uniform(-180, 180) for x in range(500)]

for index, row in location_data.iterrows():
    
    lat = row['random_lat']
    lng = row['random_lng']
    location_data.set_value(index, 'Closest_City', citipy.nearest_city(lat, lng).city_name)
    location_data.set_value(index, 'Country', citipy.nearest_city(lat, lng).country_code)

location_data
```
